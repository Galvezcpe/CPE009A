# Loop Statement

Example #1

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
```

Example #2

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

Example #3

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
     continue
  print(x)
```

Example #4

```python
for x in "week":
  print(x)
```

The range() Function

```python
for x in range(6):
  print(x)
```

```python
for x in range(2, 6, 2):
  print(x)
```

Nested Loop

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```

Example #5

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

The Break Statement

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
     print(i)
  continue
```

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("I is no longer less than 6")
```
