```python
def termCrit(actNumber):
  rawScore = []
  totalScore = []

  for i in range(0, actNumber):
    while True:
      try:
        inputScore = float(input(f"Score: "))
        break
      except ValueError:
        print("Invalid input for score. Please enter a number.")

    while True:
      try:
        inputTotal = float(input(f"Total: "))
        break
      except ValueError:
        print("Invalid input for total. Please enter a number.")

    if inputScore > inputTotal:
      print("\nInvalid Scores.\n")

      continue
    else:
      rawScore.append(inputScore)
      totalScore.append(inputTotal)

  if not totalScore or sum(totalScore) == 0:
    return 0.0

  grade = (sum(rawScore) / sum(totalScore)) * 100
  return grade

def termCS():
  print("Hands-on Activities: ")
  num_hoa = int(input("How many hands-on activities this term? "))
  termHOA = termCrit(num_hoa)

  print("Quizzes:")
  num_quizzes = int(input("How many quizzes this term? "))
  termQuiz = termCrit(num_quizzes)

  print("Assignment: ")
  num_assignments = int(input("How many assignments? "))
  termAss = termCrit(num_assignments)

  weight_hoa = 0.5
  weight_quiz = 0.3
  weight_ass = 0.2

  class_standing_grade = (weight_hoa * termHOA) + (weight_quiz * termQuiz) + (weight_ass * termAss)
  return class_standing_grade

def termGrade():
  print("Computing for the term grade.")

  while True:
    try:
      examRaw = float(input("Exam Score: "))
      break
    except ValueError:
      print("Invalid input for exam score. Please enter a number.")

  while True:
    try:
      examTotal = float(input("Exam Total: "))
      break
    except ValueError:
      print("Invalid input for exam total. Please enter a number.")

  if examTotal == 0:
    examGrade = 0.0
  else:
    examGrade = (examRaw / examTotal) * 100

  CSgrade = termCS()

  term_final_grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return term_final_grade

def semesterGrade():
  print("Prelim Grade")
  prelimGrade = termGrade()

  print("\nMidterm Grade")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("\nFinal Grade")
  finalGrade_term = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade_term

  return finalFG



overall_semester_result = semesterGrade()
print(f"{overall_semester_result:.2f}")
```
