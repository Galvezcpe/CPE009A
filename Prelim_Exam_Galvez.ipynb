```python


Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)



Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"
```

```python
print("--- Grade Calculator ---")
subjects = []
num_subjects = int(input("Enter the number of subjects: "))

for i in range(num_subjects):
    subject_name = input(f"Enter SUBJECT: ")
    marks = float(input(f"Enter Grades for {subject_name}: "))
    subjects.append({'name': subject_name, 'marks': marks})

def assign_grade(marks):
    if 90 <= marks <= 100:
        return "A+"
    elif 80 <= marks < 90:
        return "A"
    elif 70 <= marks < 80:
        return "B+"
    elif 60 <= marks < 70:
        return "B"
    elif 50 <= marks < 60:
        return "C"
    elif 40 <= marks < 50:
        return "D"
    else:
        return "F (Fail)"

total_marks_obtained = 0
total_max_marks = num_subjects * 100
overall_fail = False

print("\n--- Subject GRADES ---")
for subject in subjects:
    marks = subject['marks']
    grade = assign_grade(marks)
    subject['grade'] = grade
    total_marks_obtained += marks
    print(f"{subject['name']}: Percentage = {marks}, Grade = {grade}")
    if grade == "F (Fail)":
        overall_fail = True


overall_score = (total_marks_obtained / total_max_marks) * 100 if total_max_marks > 0 else 0

print(f"\nTotal Marks Obtained: {total_marks_obtained}")
print(f"Overall Percentage: {overall_score:.2f}%")

if overall_fail or overall_score < 40:
    overall_result = "Failed"
else:
    overall_result = "Passed"

print(f"Overall Result: {overall_result}")

overall_score == overall_result
print(f"Percentage Remarks: {performance_result}")

performance_result = ""
if overall_score > 90:
    performance_result = "Outstanding"
elif overall_score > 80:
    performance_result = "Excellent"
elif overall_score > 70:
    performance_result = "Very Good"
elif overall_score > 60:
    performance_result = "Good"
elif overall_score > 40:
    performance_result = "Satisfactory"
else:
    performance_result = "Needs Improvement"
```
